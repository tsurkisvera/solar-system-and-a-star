```python
from bodies import Solar_system
import numpy as np
```

```python
my_system = Solar_system(names=np.array(['Sun', 'Jupyter', 'Earth']), masses=np.array([1000., 1., 1./318]), initial_positions=np.array([(0., 0., 0.), (52., 0, 0), (10., 0., 0.)]), initial_velocities=np.array([(0., 0., 0.), (0., 31.5/np.sqrt(10)/np.sqrt(5.2), 0), (0., 31.5/np.sqrt(10), 0)]), time=1000., time_step=.1)
```

```python
my_system.add_body(name='Saturn',
                   mass=0.3,
                   initial_position=np.array([[101., 0., 0.]]),
                   initial_velocity=np.array([[0., 31.5/np.sqrt(10)/np.sqrt(10.1), 0.]]))

my_system.add_body(name='Neptune',
                   mass=1/318*17.29,
                   initial_position=np.array([[300.7, 0., 0.]]),
                   initial_velocity=np.array([[0., 31.5/np.sqrt(10)/np.sqrt(30.07), 0.]]))

my_system.add_body(name='Star',
                   mass=1000.,
                   initial_position=np.array([[1000, 0., 200.]]),
                   initial_velocity=np.array([[-5., 0., -1.1]]))

my_system.add_body(name='Venus',
                   mass=1/318*0.815,
                   initial_position=np.array([[7.2, 0., 0.]]),
                   initial_velocity=np.array([[0., 31.5/np.sqrt(10)/np.sqrt(.72), 0.]]))

my_system.add_body(name='Mars',
                   mass=1/318*0.107,
                   initial_position=np.array([[15.2, 0., 0.]]),
                   initial_velocity=np.array([[0., 31.5/np.sqrt(10)/np.sqrt(1.52), 0.]]))

my_system.add_body(name='Mercury',
                   mass=1/318*0.0553,
                   initial_position=np.array([[3.9, 0., 0.]]),
                   initial_velocity=np.array([[0., 31.5/np.sqrt(10)/np.sqrt(.39), 0.]]))

my_system.add_body(name='Uranus',
                   mass=1/318*14.59,
                   initial_position=np.array([[191.9, 0., 0.]]),
                   initial_velocity=np.array([[0., 31.5/np.sqrt(10)/np.sqrt(19.19), 0.]]))
```

```python
my_system.account_for = 'Sun_and_star'
```

```python
my_system.do_calculations()
```

```python
np.save('time_history_direct.npy', my_system.history.t)
```

```python
def number_of_periods(arr):
    count = 0
    for i, element in enumerate(arr[:-1]):
        if arr[i]<0 and arr[i+1]>=0:
            count += 1
    return count
```

```python
number_of_periods(my_system.history.y[2])
```

```python
n_bodies = len(my_system.names)
n_steps = my_system.history.y.shape[1]
new_positions = np.zeros((n_steps, n_bodies, 3))
for i in range(n_steps):
    for j in range(n_bodies):
        new_positions[i, j, 0] = my_system.history.y[j, i]
        new_positions[i, j, 1] = my_system.history.y[n_bodies+j, i]
        new_positions[i, j, 2] = my_system.history.y[2*n_bodies+j, i]
```

```python
np.save('positions_history_direct.npy', new_positions)
```
